```python
import glob
import concurrent.futures
import pandas as pd

MAX_CPU = 10
```

# Processing

```python
dates = ["Apr2020", "Apr2021", "Aug2020", "Aug2021", "Dec2020", "Dec2021", "Feb2020", "Feb2021", "Jan2020", "Jan2021", "Jul2020", "Jul2021", "Jun2020", "Jun2021", "Mar2020", "Mar2021", "May2020", "May2021", "Nov2020", "Nov2021", "Oct2020", "Oct2021", "Sep2020", "Sep2021"]
```

```python
df_total = []

for date in dates:
    print(date)
    logfiles = glob.glob(f"summit-data-pseudonymized/{date}/*.txt")
    #print(logfiles)
    with concurrent.futures.ProcessPoolExecutor(max_workers=MAX_CPU) as e:
        jobs = [e.submit(extract_w_block, logf) for logf in logfiles]
    #
    dfs = [me.result() for me in jobs]
    #print(dfs)
    #
    df_month = pd.concat(dfs)
    df_month.reset_index(drop=True, inplace=True)
    df_total.append(df_month)

df_total = pd.concat(df_total)
df_total.reset_index(drop=True, inplace=True)
df_total.to_feather(f"results/w_block.arrow", compression="lz4")
```
